```python
import numpy as np
happiness_rating = (ice_cream_rating + sleeping_rating) / 2
```

#### Problem 4

```python
def check_type(variable):
    print(type(variable))

check_type(ice_cream_rating)
check_type(first_name)
check_type(happiness_rating)
```

- Q: Did all of the data types make sense?
- A: Yes.

- Q: Were there any data types that were different than you expected?
- A: At first, ice_cream_rating and sleeping_rating is int type, but after these two int variables are summed and divided by 2, this new variable (happiness_rating) becomes a float variable.
